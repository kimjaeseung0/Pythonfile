```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('C:/2024_2/dataset/2017PurchasePricesDec.csv')  # Ensure the file path and name are correct

# Convert 'Volume' and 'Price' columns to numeric, forcing errors to NaN
data['Volume'] = pd.to_numeric(data['Volume'], errors='coerce')
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

# Drop rows where either 'Volume' or 'Price' is NaN
data = data.dropna(subset=['Volume', 'Price'])

# Calculate 'Total Price' as Volume * Price
data['TotalPrice'] = data['Volume'] * data['Price']

# Sort the DataFrame by 'TotalPrice' in descending order
df_sorted = data.sort_values(by='TotalPrice', ascending=False)

# Calculate the count for the top 20%
top_20_percent_count = int(len(df_sorted) * 0.2)
top_20_percent_df = df_sorted.head(top_20_percent_count)

# Calculate the sum of 'TotalPrice' for the top 20%
top_20_price_sum = top_20_percent_df['TotalPrice'].sum()

# Calculate the total sum of 'TotalPrice'
total_price_sum = df_sorted['TotalPrice'].sum()

# Calculate the percentage
percentage = top_20_price_sum / total_price_sum

# Print results
print(f"상위 20%의 가격 합계: {top_20_price_sum}")
print(f"전체 가격 합계: {total_price_sum}")
print(f"상위 20%가 차지하는 비율: {percentage * 100:.2f}%")

# Check if the top 20% is close to 80%
if percentage >= 0.8:
    print("상위 20%가 전체 가격의 80% 이상을 차지합니다.")
else:
    print("상위 20%가 전체 가격의 80% 미만을 차지합니다.")

# Visualization
labels = ['TOP 20%', 'Others']
sizes = [top_20_price_sum, total_price_sum - top_20_price_sum]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  # Explode the top 20% slice

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle
plt.title('2017PurchasePricesDec')
plt.show()
```
